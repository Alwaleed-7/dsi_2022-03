```python
def to_kg(x):
    return round(x/2.2)
```

```python
# To convert 1 lb  = 1/2.2 kg
cars["weight_kg"] = cars.weight.map(to_kg)
```

```python
cars.drop("wight_kg", axis=1, inplace=True)
```

```python
cars
```

```python
cars.weight_kg.plot.hist();
```

```python
def is_heavy(x):
    if x > 1800:
        result = "heavy"
    elif x > 1200:
        result = "medium"
    else:
        result = "light"
    return result
```

```python
cars["is_heavy"] = cars.weight_kg.apply(is_heavy)
```
